```python
#import the dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
import time
# Set up Splinter
browser = Browser('chrome')
```

```python
# define the url to inspect
url = 'https://www.imdb.com/list/ls008957859/'
#visit the browser
browser.visit(url)
last_height = browser.execute_script("return document.body.scrollHeight")
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    time.sleep(1)

    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
# parse the html
html = browser.html
song_soup = soup(html , 'html.parser')
```

```python
# get the main mage content by inspecting the class
title_elem = song_soup.find('div',  class_ = 'lister list detail sub-list')
#print(soup.prettify(title_elem))
```

```python
# get the list contating all shows details
title_list = title_elem.find_all('div',class_ = 'lister-item-content')
print(len(title_list))
```

```python
print(title_list[0])
```

```python
# create empty list
all_details_list = []
```

```python
# loop to get all the details for all the shows 
for each in title_list:
    try:
        
        titles = each.h3.a.get_text()
        title_link = each.h3.a['href'] 
        rank = each.h3.find('span', class_ = 'lister-item-index unbold text-primary').get_text().replace('.', '')
        run_years = each.h3.find('span',class_ = 'lister-item-year text-muted unbold').get_text()
        p_tag = each.find_all('p',class_ = 'text-muted text-small')
        certificate = p_tag[0].find('span', class_= "certificate").get_text().replace('TV-','')
        runtime = p_tag[0].find('span', class_= "runtime").get_text().replace('min','')
        genre = p_tag[0].find('span', class_= "genre").get_text().strip()
        Rating = each.find('span', class_ = 'ipl-rating-star__rating').get_text()
        Actor = p_tag[1].find_all('a')
        Actors = [actor.get_text() for actor in Actor]
        Votes = p_tag[2].find_all('span')[1].get_text().replace(',','')
        Synopsis = each.find_all('p')[1].get_text().strip()
    except :
        pass
    dict_elements = {'Title' : titles,
                    'url' : title_link,
                    'Rank' : rank,
                    'Years of running' : run_years,
                    'Certificate' : certificate,
                    'Run time in minutes' : runtime,
                    'Genre' : genre,
                    'Rating' : Rating,
                    'Actors' : Actors,
                    'Votes' : Votes,
                    'Synopsis' : Synopsis}                        
    # Add the dictionary to the list
    all_details_list.append(dict_elements) 
    print(dict_elements)
    
 

```

```python
print(len(all_details_list))
```

```python
# create dataframe
import pandas as pd
Shows_df = pd.DataFrame(all_details_list)
Shows_df.head(50)
```

```python
#saving data
import os  
os.makedirs('ETL_practice/Resources', exist_ok=True)  
IMDB_df = Shows_df.to_csv('ETL_practice/Resources/Top 100 TV.csv')  
```
