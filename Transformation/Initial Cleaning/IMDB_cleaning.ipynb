```python
# import dependencies
import pandas as pd
import os
```

```python
#define paths
file1 = "Resources/Resources_IMDB_data_Raw Data/Top 250_1.csv"
file2 = "Resources/Resources_IMDB_data_Raw Data/Top 250_2.csv"
file3 = "Resources/Resources_IMDB_data_Raw Data/Top 250_3.csv"
```

```python
path_list = [file1, file1, file3]
```

```python
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df3
```

```python
final_df = pd.concat([ df1, df2, df3])
final_df.keys()
```

```python
final_df = final_df.rename(columns = {'Title' : 'Link', 'url' : "Titles" })
del final_df['Unnamed: 0']
final_df
```
